```python
# Define a function to categorize based on TYPEQU
def categorize_typequ(code):
    if isinstance(code, str):
        if code.startswith('A'):
            return 'Services for Individuals'
        elif code.startswith('B'):
            return 'Shops'
        elif code.startswith('C'):
            return 'Education'
        elif code.startswith('D'):
            return 'Health and Social Action'
        elif code.startswith('E'):
            return 'Transport and Travel'
        elif code.startswith('F'):
            return 'Sports, Leisure, and Recreation'
        elif code.startswith('G'):
            return 'Tourism'
    return 'Other'
```

```python
# Apply the function to create a new column 'Main_Category'
BPE['Main_Category'] = BPE['TYPEQU'].apply(categorize_typequ)
```
